```python
#The data is split into two dataframes - numerical and categorical data
```

# Removing outliers

```python
for i in num:
    print(i)
    sns.boxplot(data=df[i])
    plt.show()
```

```python
num = df.select_dtypes(include=['float64', 'int64']).columns

for i in num:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)

    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    up = Q3 + 1.5 * IQR

    for j in df[i]:
        if j <= low:
            df = df.replace(j, low)
        if j >= up:
            df = df.replace(j, up)
```

# Encoding the categorical columns


```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for i in cat:
    cat[i] = encoder.fit_transform(df[i])
```

```python
cat
```

# Scaling numerical columns

```python
from sklearn.preprocessing import MinMaxScaler

# Select numeric columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Exclude non-numeric columns
num_data = df[num_cols]

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the numeric data
scaled_data = scaler.fit_transform(num_data)

# Create a DataFrame with the scaled data and original column names
scaled_dataset = pd.DataFrame(scaled_data, columns=num_cols)
```

```python
# Concatenate both encoded categorical dataframe and scaled numerical dataframe.

df = pd.concat([cat, scaled_dataset], axis=1)
```

```python
df
```

# model building

```python
# Display the column names to verify if 'price' is present
print(df.columns)

# Splitting the dataset
from sklearn.model_selection import train_test_split

# Assuming 'price' is a column in your DataFrame
x = df.drop(['Selling_Price'], axis=1)
y = df['Selling_Price']

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.3)
```

```python
# Splitting the dataset

from sklearn.model_selection import train_test_split

x = df.drop(['Present_Price'], axis=1)
y = df['Present_Price']

X_train,X_test, y_train, y_test  = train_test_split(x,y,random_state=42, test_size=0.3)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train, y_train)

y_pred=model.predict(X_test)
```

# Evaluating the performance of the model

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Training Accuracy: ', model.score(X_train, y_train))
print('-----------------------------------------------------')
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print('-----------------------------------------------------')
print("R-squared:", r2_score(y_test, y_pred))
```
